```python
# Batch Inference Pipeline
import pickle
import pandas as pd

# 1. Definir las etiquetas
labels = ['Anxiety', 'Bipolar', 'Depression', 'Normal', 'Personality disorder', 'Stress', 'Suicidal']

# 2. Cargar el modelo entrenado desde MLflow o un archivo local
model_path = "C:/Users/megag/Documents/Curso IA/Curso Avanzado/Machine Learning Clasification/sentiment_analysis_model.pkl"
vectorizer_path = "C:/Users/megag/Documents/Curso IA/Curso Avanzado/Machine Learning Clasification/tfidf_vectorizer.pkl"

with open(model_path, "rb") as f:
    model = pickle.load(f)

with open(vectorizer_path, "rb") as f:
    vectorizer = pickle.load(f)

# 3. Cargar nuevos datos para predicción
new_data = ["I feel very anxious today", "I'm so happy and full of energy!", "I feel like everything is falling apart"]

# 4. Vectorizar el nuevo texto
new_data_tfidf = vectorizer.transform(new_data)

# 5. Hacer predicciones
predictions = model.predict(new_data_tfidf)

# 6. Mostrar predicciones
for text, pred in zip(new_data, predictions):
    print(f"Texto: {text} -> Predicción: {pred}")

# 7. Guardar predicciones en un archivo
predictions_path = "C:/Users/megag/Documents/Curso IA/Curso Avanzado/Machine Learning Clasification/predictions.csv"
pd.DataFrame({"text": new_data, "prediction": predictions}).to_csv(predictions_path, index=False)

print("Predicciones guardadas en:", predictions_path)
```
